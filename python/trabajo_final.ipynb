# Abriendo una página web



```python
# requests es para obtener una pagina web
import requests

# Cargando BeautifulSoap para obtener la información almacenada en la página
from bs4 import BeautifulSoup

# nos permite realizar busquedas a travez de expresiones regulares
import re
```

```python
url = 'https://www.hites.com/deportes/aire-libre/piscinas-y-accesorios/'

page = requests.get(url)

page = BeautifulSoup(page.content, 'html.parser')
print(page.prettify())
```

## Obteniendo datos desde la página

### Forma solo tag

```python
# Obteniendo el primer elemento de un tag definido (título de la sección)
h1_tag = page.h1

# Viendo que tiene la variable h1_tag 
print(f'h1_tag:\n{h1_tag}')
```

```python
# Obteniendo el texto que contiene el tag h1 (texto del título de la sección)
# Para esto a la variable que contiene la información se le agrega .string
print(f"viendo texto de h1_tag:\n{h1_tag.string}")
```

### Forma find

```python
# buscando por tag usando el find, este devuelve el primer elemento que encuentra (título de la sección)
first_h1 = page.find('h1')
print(first_h1)

# obteniendo sólo el texto (texto del título de la sección)
print(first_h1.string)


# Buscando el logo de Hites con find
class_seccion = page.find(class_='header__logo')
print(class_seccion)


# Buscando class header con find
header = page.find(class_="hites-header")
print(header)

# Buscando class, ID body con find
catalog_body = page.find(id="main", class_="catalog__body")
print(catalog_body)

# Buscando ID footer con find
footer = page.find(id="footercontent")
print(footer)
```

### Forma find_all

```python
all_products = page.find_all(class_='col-6 col-lg-4 px-0')
print(len(all_products))

for product in all_products:
    # obteniendo el nombre
    name_product = product.find(class_='link product-name--bundle')
    print(name_product.string)

    #obteniendo el precio
    price_product = product.find(class_='value')
    if price_product.string is None:
        print("este producto no tiene oferta")
        price = price_product.find(text=re.compile("\$"))
        print(int(price.replace('\n','').replace(' ','').replace('Oferta','').replace('$','').replace('.','')))
    else:
        print(int(price_product.string.replace('\n','').replace(' ','').replace('Oferta','').replace('$','').replace('.','')))
```

## Obteniendo los links

```python
all_a = page.find_all('a',href=True)

for a in all_a:
    print(a['href'])
```
